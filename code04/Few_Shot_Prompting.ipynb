```python
med_url_dict = {}
for i in range(context.shape[0]-5):
    med_url_dict[context.iloc[i, 4]] = context.iloc[i,2]

for med, url in med_url_dict.items():
    # Send a GET request
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()

        file_path = f'json_files/{med}_data.json'
        with open(file_path, 'w') as json_file:
            json.dump(data, json_file, indent=4)  # Use indent to format the JSON for readability

        print(f"Data has been saved to {file_path}")
    else:
        print(f"Error: {response.status_code}")
```

```python
med_url_dict = {}
med_monography_dict = {}

med_url_dict = {}
for i in range(context.shape[0]-5):
    med_url_dict[context.iloc[i, 4]] = context.iloc[i,2]

for med, url in med_url_dict.items():
    # Send a GET request
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()

        # Retrieve relevant drug monography information and concatenate
        dosage_and_administration = data['results'][0]['dosage_and_administration'][0]
        pregnancy = data['results'][0]['pregnancy'][0] if 'pregnancy' in data['results'][0].keys() else None
        warnings = data['results'][0]['warnings'][0] if 'warnings' in data['results'][0].keys() else None
        boxed_warning = data['results'][0]['boxed_warning'][0] if 'boxed_warning' in data['results'][0].keys() else None
        contraindication = data['results'][0]['contraindication'][0] if 'contraindication' in data['results'][0].keys() else None
        do_not_use = data['results'][0]['do_not_use'][0] if 'do_not_use' in data['results'][0].keys() else None
        combined_text = " ".join(filter(None, [boxed_warning, contraindication, do_not_use, dosage_and_administration, pregnancy, warnings]))
        med_monography_dict[med] = combined_text

    else:
        print(f"Error: {response.status_code}")
```

```python
#model_id = "Qwen/Qwen2.5-0.5B"
model_id = "suayptalha/DeepSeek-R1-Distill-Llama-3B"
#model_id = "deepseek-ai/deepseek-llm-7b-base"


print("\nInitializing model and tokenizer...")
print(
        "This might take a few minutes on first run as the model needs to be downloaded."
    )

tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)

model = AutoModelForCausalLM.from_pretrained(
        model_id,
        torch_dtype=torch.float16,
        device_map="auto") ##trust_remote_code=True, cache_dir="model_cache", Cache the model for future use


print("\nCreating pipeline...")
# Create pipeline
pipe = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
        max_new_tokens=512,
        temperature=0.1,
        top_p=0.95,
        repetition_penalty=1.1,
        do_sample=True,
    )

print("Model initialization complete!")

# Create LangChain LLM
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
import pandas as pd
import time
import re
from langchain_core.runnables import RunnableMap
from langchain_core.documents import Document

# --- Step 1: Build few-shot examples from context_train with monograph ---
few_shot_examples = ""
for _, row in few_shot_rows.head(4).iterrows():
    medication = row["medication"]

    # Get structured info
    structured_info = "\n".join([
        f"{col}: {row[col]}"
        for col in context_train.columns
        if col not in ["reason_fyi", "order_id", "open_fda_api"]
    ])

    # Retrieve real chunks using your manual retriever
    retrieved_chunks = med_monography_dict.get(medication, "")

    # Build the full example
    few_shot_examples += f"""Context (Structured Data):
{structured_info}

Context (Monograph Information):
{retrieved_chunks}

Question: Is the medication prescribed safe and why?
Answer: {row['label']}. {row['reason_fyi']}

"""

# --- Step 2: Prompt Template ---
from langchain_core.prompts import PromptTemplate

template = f"""{few_shot_examples}
### END OF FEW-SHOT EXAMPLES ###

Use the following context to answer the question.

Context (Structured Data):
{{structured_context}}

Context (Monograph Information):
{{retrieved_chunks}}

Question: Is the medication prescribed safe and why?
Answer:"""

prompt = PromptTemplate.from_template(template)

# --- Step 3: Helpers ---
def build_structured_context(row):
    return "\n".join([
        f"{col}: {row[col]}"
        for col in row.index
        if col not in ["reason_fyi", "label", "order_id", "open_fda_api"]
    ])

def get_combined_context(test_row):
    return {
        "structured_context": build_structured_context(test_row),
        "retrieved_chunks": med_monography_dict.get(test_row["medication"], "")
    }

# --- Step 4: RAG Chain ---
rag_chain = get_combined_context | prompt | llm

# --- Step 5: Answer Extractor ---
def extract_final_answer_regex(text):
    matches = re.findall(r"Question:.*?Answer:(.*?)(?=Context:|Question:|### END ###|\Z)", text, re.DOTALL)
    return matches[-1].strip() if matches else text.strip()

# --- Step 6: Run on full test set ---
output_data = []

for i in range(1):
    row = context_test.iloc[i]
    medication = row['medication']
    context = row['context']
    order_id = row['order_id']
    row_for_llm = row.drop(labels=["order_id", "open_fda_api"])

    start_time = time.time()
    result = rag_chain.invoke(row_for_llm)
    end_time = time.time()
    elapsed_time = round(end_time - start_time, 3)

    truncated = extract_final_answer_regex(result)
    retrieved_chunks = med_monography_dict.get(medication, "")

    output_data.append({
        "index": i,
        "order_id": order_id,
        "medication": medication,
        "structured_context": context,
        "retrieved_chunks": retrieved_chunks,
        "full_output": result,
        "truncated_answer": truncated,
        "time_taken_sec": elapsed_time
    })

# --- Step 7: Output as DataFrame ---
output_data_df = pd.DataFrame(output_data)
output_data_df
```
